```python
#detect outlier

q3 = df_lifeStyle['alcohol_consumption_per_week'].quantile(0.75)
q1 = df_lifeStyle['alcohol_consumption_per_week'].quantile(0.25)

iqr =q3 - q1

upper = q3 + (1.5*iqr)
print(upper)
df_lifeStyle[df_lifeStyle['alcohol_consumption_per_week'] > upper].sort_values('alcohol_consumption_per_week',ascending=False)
```

```python
#detect outlier

q3 = df_lifeStyle['physical_activity_minutes_per_week'].quantile(0.75)
q1 = df_lifeStyle['physical_activity_minutes_per_week'].quantile(0.25)

iqr =q3 - q1

upper = q3 + (1.5*iqr)
lower= q1 - (1.5*iqr)
print(upper)
print(lower)
df_lifeStyle[
    (df_lifeStyle['physical_activity_minutes_per_week'] > upper)
].sort_values('physical_activity_minutes_per_week', ascending=False)
```

```python
#detect outlier

q3 = df_lifeStyle['screen_time_hours_per_day'].quantile(0.75)
q1 = df_lifeStyle['screen_time_hours_per_day'].quantile(0.25)

iqr =q3 - q1

upper = q3 + (1.5*iqr)
lower= q1 - (1.5*iqr)
print(lower)
print(upper)
df_lifeStyle[df_lifeStyle['screen_time_hours_per_day'] > upper].sort_values('screen_time_hours_per_day',ascending=False)
```

## Insights
##### 1 - Smoking is generally not linked to diabetes.
##### 2 - Lack of physical activity and excessive screen time contribute to the development of diabetes, and certainly negatively impact BMI.
##### 3 - Those with diabetes have a weight that is 1 to 1.5 times higher on the BMI scale.
##### 4 - All ethnicities have roughly equal rates of diabetes, which does not support the idea that ethnicity is related to the disease.
##### 5 - There are slight differences between those infected and those not infected in LDL, triglyceride, and HDL levels that are almost constant.
##### 6 - Systolic blood pressure is only slightly affected by diabetes, while it is greatly affected by increasing age. As for heart rate, those with a low heart rate tend to have a higher percentage of diabetic patients.
##### 7 -The hemoglobin level in the blood clearly and explicitly affects diabetes.
##### 8 -A decrease in blood insulin levels makes a person susceptible to developing diabetes.
##### 9 - Risk score is a strong indicator but should be combined with other features (like glucose and BMI) for better prediction.
